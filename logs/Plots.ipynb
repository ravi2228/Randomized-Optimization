```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['font.size'] = 12
```

```python
def plot_MSE(train_sizes, train_mean, cv_mean, title):
    
    plt.figure()
    plt.title("Error Curve: "+ title)
    plt.xlabel("Number of Iterations")
    plt.ylabel("Mean Squared Error (MSE)")
    plt.plot(train_sizes, train_mean, color="b", label="Train Error")
    plt.plot(train_sizes, cv_mean, color="r", label="Test Error")
    plt.legend(loc="best")
    plt.show()
```

```python
def plot_time(train_sizes, train_mean, cv_mean, title):
    
    plt.figure()
    plt.title("Modeling Time: "+ title)
    plt.xlabel("Number of Iterations")
    plt.ylabel("Time (s)")
    plt.plot(train_sizes, train_mean, color="b", label="Training Time")
    plt.plot(train_sizes, cv_mean, color="r", label="Prediction Time")
    plt.legend(loc="best")
    plt.show()
```

## Backpropagation

```python
df = pd.read_csv('BACKPROP_LOG.csv')
plot_MSE(df['iteration'],df['MSE_trg'],df['MSE_tst'],'Using Backpropagation Algorithm on Bank Marketing Data')
plot_time(df['iteration'],df['train_time'],df['pred_time'],'Using Backpropagation Algorithm on Bank Marketing Data')
```
